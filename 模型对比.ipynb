```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_table('./accloss/acc-loss-alexnet.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df2=pd.read_table('./accloss/acc-loss-pre-vgg.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df3=pd.read_table('./accloss/acc-loss-vgg16.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df4=pd.read_table('./accloss/acc-loss-resnet.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df5=pd.read_table('./accloss/acc-loss-pre-inception.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df6=pd.read_table('./accloss/acc-loss-mobilenet.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df7=pd.read_table('./accloss/acc-loss-inception.txt',names=['loss','acc','valloss','valacc'],sep=' ')
```

```python
df1
```

```python
acc_alexnet=list(np.array(df1['acc']))
acc_vgg16=list(np.array(df3['acc']))
acc_inception=list(np.array(df7['acc']))
acc_mobilenet=list(np.array(df6['acc']))
acc_pre_vgg=list(np.array(df2['acc']))
acc_resnet=list(np.array(df4['acc']))
acc_pre_inception=list(np.array(df5['acc']))
```

```python
from matplotlib import pyplot as plt
```

```python
epochs   = range(len(acc_alexnet))
plt.plot(epochs,acc_alexnet, label='alextnet')
plt.plot(epochs,acc_vgg16, label='vgg16')
plt.plot(epochs,acc_inception, label='inception')
plt.plot(epochs,acc_mobilenet, label='mobilenet')
plt.plot(epochs,acc_pre_vgg, label='pre-train-vgg')
plt.plot(epochs,acc_resnet, label='resnet')
plt.plot(epochs,acc_pre_inception, label='pre-train-inception')
plt.title('Model contrast--train_acc')
plt.legend()
```

## acc

```python
valacc_alexnet=list(np.array(df1['valacc']))
valacc_vgg16=list(np.array(df3['valacc']))
valacc_inception=list(np.array(df7['valacc']))
valacc_mobilenet=list(np.array(df6['valacc']))
valacc_pre_vgg=list(np.array(df2['valacc']))
valacc_resnet=list(np.array(df4['valacc']))
valacc_pre_inception=list(np.array(df5['valacc']))
```

```python
epochs   = range(len(acc_alexnet))
plt.plot(epochs,valacc_alexnet, label='alextnet')
plt.plot(epochs,valacc_vgg16, label='vgg16')
plt.plot(epochs,valacc_inception, label='inception')
plt.plot(epochs,valacc_mobilenet, label='mobilenet')
plt.plot(epochs,valacc_pre_vgg, label='pre-train-vgg')
plt.plot(epochs,valacc_resnet, label='resnet')
plt.plot(epochs,valacc_pre_inception, label='pre-train-inception')
plt.title('Model contrast--test_acc')
plt.legend()
```
